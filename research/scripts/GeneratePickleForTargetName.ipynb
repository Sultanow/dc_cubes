```python
def createPickleForOneTargetName(targetName, pickleName):
    import os
    files = os.listdir('C:/Users/Sebastian/Desktop/Uni/_Bachelorarbeit/data/chunks')
    path = 'C:/Users/Sebastian/Desktop/Uni/_Bachelorarbeit/data/chunks/chunk'
    upperRangeLimit = len(files) # doesn't need +1 because i have one folder in there, if there are only the chunk files take len+1
    frame = pd.DataFrame()
    for counter in range(1, upperRangeLimit):
        currentFile = path + str(counter) + ".pkl"
        print("file: " + currentFile)
        with(open(path+str(counter)+".pkl", "rb")) as pklfile:
            currentchunk = pickle.load(pklfile)
            data = currentchunk.loc[currentchunk.TARGET_NAME == targetName]
            #print(df_filtered)
            if not data.empty:
                frame = pd.concat([frame,data])
            else:
                print("empty")
    with open(pickleName, "wb") as pickleFile:
            pickle.dump(frame, pickleFile)        
```
